## **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Gemini said
Penggunaan Median Imputation pada kolom salary_offered dengan nilai 5.115 merupakan langkah strategis untuk menjaga integritas data keuangan tanpa terdistorsi oleh nilai-nilai ekstrem atau pencilan yang sering muncul dalam struktur penggajian. Secara teknis, pengisian nilai kosong menggunakan median memastikan bahwa distribusi data tetap stabil dan representatif terhadap kondisi pasar kerja yang sebenarnya, karena median cenderung lebih kebal terhadap ketimpangan antara gaji posisi entry-level dan posisi eksekutif dibandingkan nilai rata-rata. Dengan menetapkan angka 5.115 sebagai standar bagi data yang hilang, model analisis kini memiliki basis perhitungan yang konsisten, di mana lowongan-lowongan tanpa keterangan gaji dianggap memiliki penawaran kompensasi yang moderat dan berada di titik tengah persaingan industri. Secara keseluruhan, langkah ini membantu meminimalisir bias dalam pemodelan prediktif, sehingga setiap baris data dapat memberikan kontribusi yang lebih akurat dalam menggambarkan tren nilai ekonomi di dalam dataset tersebut.

```python
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np

df_knn = df.copy()

if df_knn['salary_offered'].dtype == 'object':
    df_knn['salary_offered'] = pd.to_numeric(df_knn['salary_offered'], errors='coerce')
median_salary_offered = df_knn['salary_offered'].dropna().median()
df_knn['salary_offered'] = df_knn['salary_offered'].fillna(median_salary_offered)


df_knn['total_applicants'] = pd.to_numeric(df_knn['total_applicants'].astype(str).str.replace(r'\\D', '', regex=True), errors='coerce')

median_total_applicants = df_knn['total_applicants'].dropna().median()
df_knn['total_applicants'] = df_knn['total_applicants'].fillna(median_total_applicants)

median_days_to_fill = df_knn['days_to_fill'].dropna().median()
df_knn['days_to_fill'] = df_knn['days_to_fill'].fillna(median_days_to_fill)

edu_mapping = {'SMA/SMK': 0, 'D3': 1, 'S1': 2, 'S2/S3': 3}
inv_edu_mapping = {v: k for k, v in edu_mapping.items()}
df_knn['education_required_num'] = df_knn['education_required'].map(edu_mapping)

cols_for_knn_features = ['salary_offered', 'total_applicants', 'education_required_num']

imputer = KNNImputer(n_neighbors=3)
df_imputed_numeric_for_edu = pd.DataFrame(imputer.fit_transform(df_knn[cols_for_knn_features]), columns=cols_for_knn_features)

df_knn['education_required_imputed'] = df_imputed_numeric_for_edu['education_required_num'].round().map(inv_edu_mapping)

df['salary_offered'] = df_knn['salary_offered']
df['total_applicants'] = df_knn['total_applicants'].astype('Int64')
df['education_required'] = df_knn['education_required_imputed']
df['days_to_fill'] = df_knn['days_to_fill']
```

Penggunaan KNNImputer untuk kolom education_required memberikan pendekatan yang jauh lebih presisi dibandingkan imputasi sederhana, karena ia bekerja dengan memahami konteks hubungan antar-variabel. Dengan memetakan kualifikasi pendidikan ke dalam skala numerik, algoritma ini mampu memprediksi syarat pendidikan yang hilang berdasarkan kemiripan pola pada kolom salary_offered dan total_applicants. Secara teknis, proses ini memastikan bahwa data yang dihasilkan tetap logis; misalnya, sebuah lowongan dengan gaji tinggi dan jumlah pelamar yang kompetitif akan cenderung diprediksi memiliki syarat pendidikan S1 atau S2/S3, sesuai dengan "tetangga" data yang paling mirip. Integrasi antara pengisian median pada metrik waktu dan KNN pada metrik kualifikasi ini menciptakan dataset yang sangat harmonis, di mana setiap informasi yang hilang tidak hanya diisi secara acak, melainkan disesuaikan dengan realita ekonomi dan tren rekrutmen yang ada di dalam dataset tersebut.

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

df_knn = df.dropna(subset=['salary_offered', 'total_applicants', 'gender_required'])

X = df_knn[['salary_offered', 'total_applicants']]
y = df_knn['gender_required']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)


knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)


score = knn.score(X_test, y_test)
print(f'Akurasi model KNN (Target: Gender): {score:.2f}')
```

Penerapan Median Imputation pada kolom total_applicants dengan nilai 127.0 merupakan langkah yang sangat efektif untuk menjaga objektivitas data popularitas lowongan kerja tanpa terganggu oleh angka-angka ekstrem. Dalam dunia rekrutmen, jumlah pelamar bisa sangat bervariasi; beberapa posisi sangat populer dengan ribuan pelamar, sementara posisi spesialis mungkin hanya memiliki sedikit pelamar. Dengan menggunakan median sebagai nilai pengganti data yang hilang, kita menetapkan bahwa lowongan tersebut memiliki tingkat daya tarik yang "normal" atau rata-rata di mata pencari kerja. Secara teknis, hal ini mencegah algoritma KNN memberikan bobot yang salah (bias) akibat adanya nilai pelamar yang terlalu tinggi (outlier) yang biasanya akan menarik nilai rata-rata secara tidak proporsional. Hasilnya, distribusi data tetap terjaga konsistensinya, memungkinkan model untuk membandingkan tingkat kompetisi antar lowongan secara lebih adil dan akurat.

##**Outliers**

Outliers adalah nilai data yang angkanya jauh lebih besar atau lebih kecil dibandingkan sebagian besar data lainnya. Sederhananya, outlier itu data yang “menyimpang” dari pola umum. Pada tahap ini kita harus mengetahui dulu outliers yang ada pada satu data terjadi dikarenakan hal apa, lalu kita putuskan apakah perlu dihapus, dibiarkan, atau ditransformasi agar tidak mempengaruhi hasil analisis.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Meskipun kolom days_to_fill menunjukkan angka pencilan yang cukup tinggi (12.5%), keputusan untuk tetap mempertahankan data tersebut diambil karena outliers dalam dataset rekrutmen mencerminkan kondisi riil di lapangan. Dalam dunia nyata, sangat wajar jika sebuah posisi spesialis tingkat tinggi membutuhkan waktu pengisian yang ekstrem lama, atau sebaliknya, posisi mendesak terisi dengan sangat cepat; menghapus data ini justru akan menghilangkan informasi berharga tentang variabilitas proses bisnis.
